```python

#---------------------------------@DevTown------------------------------------------------------
#-----------------MACHINE_LEARNING_BOOTCAMP_07DAYS_FINAL_PROJECT--------------------------------
#-----MOHD RAZZAK------------------------------------------------
import numpy as np # np use  as a stardard name
import pandas as pd  

from sklearn.metrics import r2_score #   r2 score for a model which predicts mean value always is 0; 
```

```python
from sklearn.datasets import load_boston #for load boston datasfrom sklearn.datasets 
boston_dataset = load_boston() 
```

```python
print(boston_dataset.keys())  #we will have to print values of bosto_dataset by keys method
```

```python
print(boston_dataset.DESCR)  #DESCR fpr describes the dataset
```

```python
boston = pd.DataFrame(boston_dataset.data, columns=boston_dataset.feature_names)
boston.head(8) # for print first 8 data
```

```python
boston['MEDV'] = boston_dataset.target
boston.head(8)  # target value MEDV is missing from the data We create a new column of target values 
```

```python
X = pd.DataFrame(np.c_[boston['LSTAT'], boston['RM']], columns = ['LSTAT','RM'])
Y = boston['MEDV']

boston.head(8)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=5)    

print(X_train.shape)  # shape for checking shape of train

print(X_test.shape)  # shape for checking shape of test

print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lin_model = LinearRegression()
lin_model.fit(X_train, Y_train)

# note : we use LinearRegression to train our model on both the training and test dataset
```

```python
#  for training set
y_train_predict = lin_model.predict(X_train)
my_rmse = (np.sqrt(mean_squared_error(Y_train, y_train_predict)))


#  ----------------here R2 score is used to evaluate the performance of a linear regression model.
my_r2 = r2_score(Y_train, y_train_predict)   


print("==============================================================================================")
print(f'Razzak_RMSE is {my_rmse}') # i have use  F string  method for printing the value
print(f'Razzak_R2_score is {my_r2}\n')


# for testing set
y_test_predict = lin_model.predict(X_test)
my_rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))
my_r2 = r2_score(Y_test, y_test_predict)

print("================================================================================================")
print(f'Razzak_RMSE is {my_rmse}') 
print(f'Razzak_R2_score is {my_r2}')
```
